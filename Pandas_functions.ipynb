### **Series object attributes**
The Series attribute is defined as any information related to the Series object such as size, datatype. etc.

**Series.index	Defines the index of the Series.**

**Series.shape	It returns a tuple of shape of the data.**

**Series.dtype	It returns the data type of the data.**

**Series.size	It returns the size of the data.**

**Series.empty	It returns True if Series object is empty, otherwise returns false.**

**Series.hasnans	It returns True if there are any NaN values, otherwise returns false.**

**Series.nbytes	It returns the number of bytes in the data.**

**Series.ndim	It returns the number of dimensions in the data.**




```python
import pandas as pd
PySer=pd.Series(data=[np.nan,18.6,22.5], index=['a','b','c'])
print(PySer.index)
print(PySer.values)
print(PySer.shape)
print(PySer.dtype)
print(PySer.size)
print(PySer.empty)
print(PySer.hasnans)#if incase np.nan is present instaed of a any datatype in the series if retiurns true otherwise false.
print(PySer.nbytes)
print(PySer.ndim)
```

**Replace Using Mean, Median, or Mode**
A common way to replace empty cells, is to calculate the mean, median or mode value of the column. Pandas uses the mean() median() and mode() methods to calculate the respective values for a specified column.

```python
import pandas as pd
df = pd.read_csv('data.csv')
x = df["Calories"].mean()
df["Calories"].fillna(x, inplace = True)
print(df.to_string())
```

**DATA FILTERING (USING CONDITIONS) IN PANDAS**
For small data sets you might be able to replace the wrong data one by one, but not for big data sets. To replace wrong data for larger data sets you can create some rules, e.g. set some boundaries for legal values, and replace any values that are outside of the boundaries.

```python
import pandas as pd
df = pd.read_csv('ddata.csv')
for x in df.index:
  if df.loc[x, "Duration"] > 120:
    df.loc[x, "Duration"] = 120
print(df.to_string())
```

```python
import pandas as pd
df = pd.read_csv('ddata.csv')
for x in df.index:
  if df.loc[x, "Duration"] > 120:
    df.drop(x, inplace = True)
print(df.to_string())
```

**Pandas has full-featured, high performance in-memory join operations idiomatically very similar to relational databases like SQL. Pandas provides a single function, merge, as the entry point for all standard database join operations between DataFrame objects**



```python
# Syntax:
# pd.merge(left, right, how='inner', on=None, left_on=None, right_on=None,
# left_index=False, right_index=False, sort=True)
```

```python
import pandas as pd
left = pd.DataFrame({
   'id':[1,2,3,4,5],
   'Name': ['Alex', 'Amy', 'Allen', 'Alice', 'Ayoung'],
   'subject_id':['sub1','sub2','sub4','sub6','sub5']})
right = pd.DataFrame(
   {'id':[1,2,3,4,5],
   'Name': ['Billy', 'Brian', 'Bran', 'Bryce', 'Betty'],
   'subject_id':['sub2','sub4','sub3','sub6','sub5']})
print(left)
print(right)
```

```python
# Example:Merge Two DataFrames on a Key
import pandas as pd
left = pd.DataFrame({
   'id':[1,2,3,4,5],
   'Name': ['Alex', 'Amy', 'Allen', 'Alice', 'Ayoung'],
   'subject_id':['sub1','sub2','sub4','sub6','sub5']})
right = pd.DataFrame({
	'id':[1,2,3,4,5],
   'Name': ['Billy', 'Brian', 'Bran', 'Bryce', 'Betty'],
   'subject_id':['sub2','sub4','sub3','sub6','sub5']})
print(pd.merge(left,right,on='id'))
```

**Merge Using 'how' Argument**
The how argument to merge specifies how to determine which keys are to be included in the resulting table. If a key combination does not appear in either the left or the right tables, the values in the joined table will be NA.




```python
# Merge Method 	SQL Equivalent 	Description
# left 		  LEFT OUTER JOIN 	  Use keys from left object
# right 		RIGHT OUTER JOIN 	  Use keys from right object
# outer 		FULL OUTER JOIN 	  Use union of keys
# inner 		INNER JOIN        	Use intersection of keys
```

```python
# Example:Left Join
import pandas as pd
left = pd.DataFrame({
   'id':[1,2,3,4,5],
   'Name': ['Alex', 'Amy', 'Allen', 'Alice', 'Ayoung'],
   'subject_id':['sub1','sub2','sub4','sub6','sub5']})
right = pd.DataFrame({
   'id':[1,2,3,4,5],
   'Name': ['Billy', 'Brian', 'Bran', 'Bryce', 'Betty'],
   'subject_id':['sub2','sub4','sub3','sub6','sub5']})
print(pd.merge(left, right, on='subject_id', how='left'))
```

```python
#Right Join
import pandas as pd
left = pd.DataFrame({
   'id':[1,2,3,4,5],
   'Name': ['Alex', 'Amy', 'Allen', 'Alice', 'Ayoung'],
   'subject_id':['sub1','sub2','sub4','sub6','sub5']})
right = pd.DataFrame({
   'id':[1,2,3,4,5],
   'Name': ['Billy', 'Brian', 'Bran', 'Bryce', 'Betty'],
   'subject_id':['sub2','sub4','sub3','sub6','sub5']})
print(pd.merge(left, right, on='subject_id', how='right',sort=True))
```

**You define a lambda function (score) that standardizes the data by subtracting the mean and dividing by the standard deviation, scaled by 10. This transformation is applied within each group (i.e., for each team).**

```python
import pandas as pd
ipl_data = {'Team': ['Riders', 'Riders', 'Devils', 'Devils', 'Kings',
   'kings', 'Kings', 'Kings', 'Riders', 'Royals', 'Royals', 'Riders'],
   'Rank': [1, 2, 2, 3, 3,4 ,1 ,1,2 , 4,1,2],
   'Year': [2014,2015,2014,2015,2014,2015,2016,2017,2016,2014,2015,2017],
   'Points':[876,789,863,673,741,812,756,788,694,701,804,690]}
df = pd.DataFrame(ipl_data)
grouped = df.groupby('Team')
score = lambda x: (x - x.mean()) / x.std()*10
print(grouped.transform(score))
```

```python
#sample()method it is used to generate random row or column from the funtion caller data frame.
import pandas as pd
data=pd.read_csv('employees.csv')
FRow=data.sample(n=1)
print(FRow)
SRow=data.sample(n=2)
print(SRow)
```

```python
#generating 25% sample of data frame
import pandas as pd
data=pd.read_csv('employees.csv')
rows=data.sample(frac=0.25)
print(rows)
```

```python
#where()method
#it is to check the data frame for one or more conditions and return result
import pandas as pd
data=pd.read_csv('employees.csv')
data.sort_values('salary',inplace=True)
filter=data['salary']=='6.0 lacs'
data.where(filter,inplace=True)
print(data)
```

```python
#pct_change()method
#it is used to find the percentage change between the current and a prior ele.
import pandas as pd
list=[10,14,20,25,12.5,13,0,50]
series=pd.Series(list)
result=series.pct_change()
print(result)
```

```python
#handling null values
#bfill stands for back fill and will fill null values with values at their very next position
import pandas as pd
import numpy as np
list=[10,np.nan,14,20,25,12.5,13,0,50]
series=pd.Series(list)
result=series.pct_change()
result=series.pct_change(fill_method='bfill')
print(result)
```

```python
#rank()method
#it returns  a rank of every respective index of a  series passed.
import pandas as pd
data=pd.read_csv('employees.csv')
data['rank']=data['salary'].rank()
print(data)
```

```python
#melt()method
#it is used to unpivot a given dataframe wide format to long format, optionally leaving identifier variables set.
import pandas as pd
df=pd.DataFrame({'name':{0:'akhi',1:'bhanu',2:'ab'},
                 'course':{0:'masters',1:'graduate',2:'graduate'},
                'age':{0:27,1:23,2:21}})
print(pd.melt(df,id_vars=['name'],value_vars=['course']))
#
```

```python
#explode()method it returns dataframe exploded lists to rows of the subset columns
#index will be duplictated for these rows
import pandas as pd
df=pd.DataFrame({'A':[[1,3],[3]],'B':1})
print(df)
print(df.explode('A'))
```

```python
import pandas as pd
df=pd.DataFrame({'A':[[1,3],[4]],'B':1})
print(df)
print(df.explode('A',ignore_index=True))
```

```python
import pandas as pd
df=pd.DataFrame({'A':[[1,2],[]],'B':1})
print(df)
print(df.explode('A',ignore_index=True))
```

```python
import pandas as pd
df=pd.DataFrame({'A':[[1,2],[]],'B':1})
print(df)
print(df.explode('A',ignore_index=False))
```

```python
#Nunique()method
#it counts the no of distinct or unique observations over the requested axis.
import pandas as pd
df=pd.DataFrame({'A':[1,2,3],'B':[1,1,1],'C':[2,5,5]})
print(df)
print(df.nunique(axis=0))
```
